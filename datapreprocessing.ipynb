```python
from bs4 import BeautifulSoup
import pandas as pd
html_file_path = 'C:/Users/user/Documents/GitHub/DS_Portfolio/lol_champvec/champlist.html'
with open(html_file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()
# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Extract the data
data = []
for row in soup.select('.rows'):
    champ_id = row.get('id')
    champ_name = row.select_one('.champImage img')['onmouseover'].split("'")[1]
    dmg = row.select_one('.dmg').text
    dmgspd = row.select_one('.dmgspd').text
    hp = row.select_one('.hp').text
    mana = row.select_one('.mana').text
    hregen = row.select_one('.hregen').text
    mregen = row.select_one('.mregen').text
    defense = row.select_one('.def').text
    mdef = row.select_one('.mdef').text
    spd = row.select_one('.spd').text
    rng = row.select_one('.rng').text

    data.append({
        'id': champ_id,
        'champ': champ_name,
        'dmg': dmg,
        'dmgspd': dmgspd,
        'hp': hp,
        'mana': mana,
        'hregen': hregen,
        'mregen': mregen,
        'def': defense,
        'mdef': mdef,
        'spd': spd,
        'rng': rng
    })

# Create a DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
# Save the DataFrame to CSV file
df.to_csv('champlist.csv', index=False, encoding='utf-8-sig')
```
